```python
# renaming columns
df_customers.rename(columns = {'First Name' : 'first_name', 'Surnam' : 'last_name', 'Gender' : 'gender', 'STATE' : 'state', 'Age' : 'age', 'n_dependants' : 'dependants_count'}, inplace = True)
```

```python
# checking for mixed columns
for col in df_customers.columns.tolist():
  weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers[weird]) > 0:
    print (col)
```

```python
# checking for null values
df_customers.isnull().sum()
```

```python
# creating subset containing null values

df_nan = df_customers[df_customers['first_name'].isnull() == True]
df_nan.head()

# note to output: I accidentally executed this code again after deleting the 'first name' column, 
# thats the reason for the error
```
