```python
# create subset containing customers younger than 25
young_parent_age = ords_prods_all[ords_prods_all['age'] < 25]
```

```python
# create new column containing most_ordered_product

young_parent_age['most_ordered_product'] = young_parent_age.groupby(['user_id'])['department_id'].transform(np.median)
```
